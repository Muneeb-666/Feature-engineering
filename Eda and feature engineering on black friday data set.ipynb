# Handling outliers

```python
#Function for outlier detection


outliers=[]

def detect_outliers(data):
    limit=3
    mean=np.mean(data)
    std=np.std(data)
    
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>limit:
            outliers.append(i)
    return outliers
```
