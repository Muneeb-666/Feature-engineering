   5.Box cox transformation
   
  Some ML algo like linear and logistic assume that the featuresa re normally distributed so they work pretty well when the 
  data is normally distributed
  they will give good accuracy and perfomance, we use these techniques are used to normally disributed the data

```python
df
```

```python
# if you want to check whether feature is normally dist or not
## Q-Q plot
import scipy.stats as stat
import pylab
```

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'Age')
```

```python
#log transformation
df['Age_log']=np.log(df['Age'])
plot_data(df,'Age_log')
```

```python
#reciprocal transformation
df['age_reciprocal']=1/df.Age
plot_data(df,'age_reciprocal')
```

```python
# square root transformation
df['age_square']=df.Age**(1/2)
plot_data(df,'age_square')
```

```python
#exponential transfoemation
df['Age_exp']=df.Age**(1/1.2)
plot_data(df,'Age_exp')
```

```python
#box cox transformation
df['age_boxcox'],parameters=stat.boxcox(df['Age'])
plot_data(df,'age_boxcox')
```
